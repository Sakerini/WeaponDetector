```python
models_list = [
    'svc',
    'knn',
    'log',
    'random_forest',
    'gaussian',
    'perceptron',
    'sgd',
    'linear_svc',
    'decision_tree'
]
```

```python
from config import *
import pandas as pd

# test_index = [30,46]
test_index = range(10)

test_df = pd.read_csv(test_processed).iloc[test_index]
x_test_original = test_df.drop(columns='Longevity')

test_df = test_df[['Education','Sex','Age','Activity','Pet','Children','Sport','Longevity']]
y_test = test_df['Longevity']
x_test = test_df.drop(columns='Longevity')
```

```python
import requests
import json

columns = ['Longevity'] + models_list

output = pd.DataFrame(columns=columns)

# host = '195.19.36.72'
host = '192.168.1.2'
port = 1111
url  = 'http://{}:{}'.format(host, port)
api_url = url + '/api'
```

```python
x_arr = x_test.to_dict(orient='records')

for i in range(len(x_arr)):
#     x_arr[i]['Model'] = models_list[2]
    r = requests.get(api_url, params=x_arr[i])
    d = json.loads(r.text.replace("'",'"'))
    d['Longevity'] = y_test[test_index[i]]
    output = output.append(d, ignore_index=True)

output
```

```python
import pickle

models = {}

# Load all models
for m in models_list:
    with open('../model/{}.pkl'.format(m), 'rb') as f:
        models[m] = pickle.load(f)
```

```python
output_pred = pd.DataFrame()
output_pred['Longevity'] = y_test
for m in models_list:
    output_pred[m] = models[m].predict(x_test_original)
```
